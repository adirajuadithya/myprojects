```python
import numpy as np
```

```python
from scipy import spatial
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
embeddings_dict = {}
```

```python
with open("glove.6B.50d.txt", 'r',encoding="utf-8") as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], "float32")
        embeddings_dict[word] = vector
```

```python
def find_closest_embeddings(embedding):
    return sorted(embeddings_dict.keys(), key=lambda word: spatial.distance.euclidean(embeddings_dict[word], embedding))
```

```python
lambda word: spatial.distance.euclidean(embeddings_dict[word], embedding)
```

```python
find_closest_embeddings(embeddings_dict["pile"])
```

```python
find_closest_embeddings(embeddings_dict["pile"])[:5]
```

```python
print(find_closest_embeddings(
    embeddings_dict["happy"] - embeddings_dict["excited"] + embeddings_dict["anxious"]
)[:5])
```

```python
tsne = TSNE(n_components=2, random_state=0)
```

```python
words =  list(embeddings_dict.keys())
vectors = [embeddings_dict[word] for word in words]
```

```python
Y = tsne.fit_transform(vectors[:1000])
```

```python
plt.scatter(Y[:, 0], Y[:, 1])
```

```python
plt.show()
```

```python
tsne = TSNE(n_components=2, random_state=0)
words =  list(embeddings_dict.keys())
vectors = [embeddings_dict[word] for word in words]
Y = tsne.fit_transform(vectors[:1000])
plt.scatter(Y[:, 0], Y[:, 1])

for label, x, y in zip(words, Y[:, 0], Y[:, 1]):
    plt.annotate(label, xy=(x, y), xytext=(0, 0), textcoords="offset points")
plt.show()
```

```python
tsne2 = TSNE(n_components=2, random_state=0)
words2 =  list(embeddings_dict.keys())
vectors2 = [embeddings_dict[word] for word in words2]
Y2 = tsne2.fit_transform(vectors2[:70])
plt.scatter(Y2[:, 0], Y2[:, 1])

for label, x, y in zip(words, Y2[:, 0], Y2[:, 1]):
    plt.annotate(label, xy=(x, y), xytext=(0, 0), textcoords="offset points")
plt.show()
```
